# 複数のCSVファイルから、一部の要素を一括して抽出する方法

オープンデータなどでデータを収集しても、ファイルがバラバラに分割されていて、
そのままでは分析に使えないケースなどがあります。
ここでは、複数のCSVファイルの内容→１つのデータフレームに集約する→１つのファイルで出力します

ちなみに、利用しているデータはe-statに収納されている政府統計データです。

```python
import numpy as np
import pandas as pd

import glob
import os
```

```python
df1 = pd.read_csv("ファイルパス/.csv", header=None, encoding="sjis")
cols = ['Q_num','taishou', 'Question','ninzuu']

df_Q = pd.DataFrame(index=[glob.glob(os.path.join('ファイルパス/.csv'))], columns=cols)

df_Q.iat[0,1] = df1.iloc[1][0]
df_Q.iat[0,2] = df1.iloc[2][0]
df_Q.iat[0,3] = df1.iloc[7][1]
df_Q
```

```python
#ファイルを一括で読み込む方法・・・ファイルパスに続いて、*.csv
files = glob.glob(os.path.join('ファイルパス/*.csv'))
files
#ファイルパスがリスト形式で格納される
```

```python
#for文を使って、フォルダ内の全ファイルを読み込む
#ただし、今回抽出するのは４カラムのみ

for file in files:
    tmp_df = pd.read_csv(file, header=None, encoding="sjis")
    
    #CSVのファイル名からQ番号を取得
    Q_num = file
    Q_num = str(Q_num)
    Q_num = Q_num[-7:]
    Q_num = Q_num[:3]
    #print(Q_num) 
    
    taishou=tmp_df.iloc[1][0]
    Question=tmp_df.iloc[2][0]
    ninzuu=tmp_df.iloc[7][1]
    df_tmp2 = pd.DataFrame([[Q_num,taishou,Question,ninzuu]],columns = cols )
    df_Q = df_Q.append(df_tmp2,ignore_index=True)
    #print(df_Q)

df_Q.head(10)
```

```python
df_Q =df_Q.drop(0)
df_Q
```

```python
df_Q.to_csv("H26質問リスト（保護者）.csv")
```
