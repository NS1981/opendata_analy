```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_csv("03input_4カ年結合_欠損など修正後.csv", header=None, encoding="utf-8")
df1.head()
```

```python
#カラム名の修正
df1_col = df1.iloc[0]
df1.columns = df1_col
df1=df1.drop([0])
df1
```

```python
#利用するカラムのリストを読み込んで、リスト化
df2 = pd.read_csv("03input_分析使用カラム.csv", header=None, encoding="utf-8")
df2 = df2.iloc[:,0]#serisに変換
list1 = df2.values.tolist()
list1
```

```python
#対象の項目だけを抽出したＤＦ
df3 = df1[list1]
df3 = df3.astype(float)#値を数値化
df3
```

```python
# 説明変数から "暴力件数""いじめ件数""不登校件数" を削除
df_setsumei = df3.drop("暴力件数", axis=1)
df_setsumei = df_setsumei.drop("いじめ件数", axis=1)
df_setsumei = df_setsumei.drop("不登校件数", axis=1)
df_setsumei.shape
```

```python
#多重共線性のチェック

df_corr = df_setsumei.corr()
df_corr.to_csv("03output_相関行列.csv")
print(df_corr)
#csvで相関行列を確認、±0.5より大きいものを確認
```

# 多重共線性を確認し、削除対象のカラムリストを作成します（手作業）

```python
#相関をチェックし削除するリストを読み込み
df_del = pd.read_csv("03input_多重共線性_削除対象カラム.csv", header=None, encoding="utf-8")
df_del
```

↑マルチコにより、削除することにしたカラム名の一覧です

```python
df_del = df_del.iloc[:,0]#serisに変換
list2 = df_del.values.tolist()#リストに変換

#対象項目を削除
for i in list2:
    df_setsumei = df_setsumei.drop(i, axis=1)
df_setsumei
```

↓ここから重回帰分析。今回は、結果を一覧で表示してくれるstatmodelsというライブラリを使用します。

```python
#重回帰分析stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

↓１暴力件数とスマホ利用の関係・・・自由度修正済み決定係数0.722

```python
#説明変数
X = df_setsumei.as_matrix()
#目的変数
Y = df3['暴力件数'].as_matrix()

results = sm.OLS( Y , X ).fit()


print(results.summary())
```

↓２いじめ認知件数とスマホ利用の関係・・・自由度修正済み決定係数0.638

```python
#説明変数
X = df_setsumei.as_matrix()
#目的変数
Y = df3['いじめ件数'].as_matrix()

results = sm.OLS( Y , X ).fit()


print(results.summary())
```

↓３不登校件数とスマホ利用の関係・・・自由度修正済み決定係数0.969　非常に高い！！！

```python
#説明変数
X = df_setsumei.as_matrix()
#目的変数
Y = df3['不登校件数'].as_matrix()

results = sm.OLS( Y , X ).fit()


print(results.summary())
```
