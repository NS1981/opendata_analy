# アンケート結果の部分抽出

01の作業で、アンケートの項目（質問文）と回答者数をリスト化しました。このリストを精査し、回答者数が十分な項目だけを抽出したいと思います。また、アンケート結果のファイルをみて頂くとわかるのですが、１つの項目に関して、さまざまな方法での集計がなされています（年齢ごとの集計や、都道府県単位での集計など）。この先の作業で、都道府県ごとの特性を分析するので、ここでは、都道府県単位の集計部分だけを抽出するします。

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import glob
import os
```

```python
df1 = pd.read_csv("作業フォルダ/インターネット利用調査（本人）H26/01.csv", header=None, encoding="sjis")

#６行目まで削除
df1 = df1.drop([0,1,2,3,4,5])
df1 
#出力結果が次の通りであればOK、861 rows × 30 columns
```

↑このように様々な集計方法が記載されているので、１つの質問のレポートが861行にもなっています。
このうち、今回使うのは都道府県単位の集計結果のみなので50行足らずですが、
この時点では、このまま残して作業を進めます。
全質問の結果を結合（横持ちデータ）したのち、都道府県以外の集計結果を削除します。

```python
#カラム名を変更、全カラムに質問番号を付与
SR_col = df1.iloc[0]
list_col = SR_col.values.tolist()

#CSVのファイル名から質問番号を取得
Q_num = glob.glob(os.path.join('作業フォルダ/インターネット利用調査（本人）H26/01.csv'))
Q_num = str(Q_num)
Q_num = Q_num[-8:]
Q_num = Q_num[:2]

for i in range(len(list_col)):
    col_name = Q_num+str(list_col[i])
    list_col[i] = col_name
    
df1.columns = list_col
df1 = df1.drop([6])
df1
#カラム名に質問番号が付与されていればOK
```

```python
#繰り返し処理を実行して、複数の質問ファイルをひとつのDFにマージ

#必要なCSVファイルのリスト
files = glob.glob(os.path.join('C:/Users/user/OneDrive/ドキュメント/DSA/06プレジョブ/作業フォルダ/インターネット利用調査（本人）H26/*.csv'))
#print(files)

for file in files:
    tmp_df = pd.read_csv(file, header=None, encoding="sjis")

    #６行目まで削除
    tmp_df = tmp_df.drop([0,1,2,3,4,5])
    
    #カラム名を変更、全カラムに質問番号を付与
    SR_col = tmp_df.iloc[0]
    list_col = SR_col.values.tolist()

    #CSVのファイル名からQ番号を取得
    Q_num = file
    Q_num = str(Q_num)
    Q_num = Q_num[-6:]
    Q_num = Q_num[:2]
    #print(Q_num) 

    for i in range(len(list_col)):
        col_name = Q_num+str(list_col[i])
        list_col[i] = col_name
    #print(list_col)

    #dfのカラムをQ番号ありに修正
    tmp_df.columns = list_col
    tmp_df = tmp_df.drop([6])
    #print(tmp_df)

    #0列目はインデックスなので削除
    tmp_df = tmp_df.drop(columns=tmp_df.columns[[0]])
    #print(tmp_df)
    
    #df1に結合
    df1 = pd.concat([df1, tmp_df], axis=1)

    
df1
#出力結果が次の通りであればOK、860 rows × 261 columns
```

```python
#読み込んだファイルの確認
print(len(files))#14ファイルであればOK
print(files)
```

↓ここで、都道府県単位以外の集計結果を削除します

```python
#都道府県別の行、以外は削除
del_list1 = list(range(8, 92, 1))
del_list2 = list(range(139, 867, 1))

df_prefecture = df1.drop(del_list1)
df_prefecture = df_prefecture.drop(del_list2)
df_prefecture
#出力結果が次の通りならOK、48 rows × 261 columns
```
